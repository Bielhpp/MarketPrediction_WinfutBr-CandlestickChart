```python
# Convertendo a coluna 'Data' para datetime e Ordenando o DataFrame pela coluna de datas
winfutcandle5min = pd.read_excel('winfut_5min.xlsx');
winfutcandle5min['Data'] = pd.to_datetime(winfutcandle5min['Data'], dayfirst=True)
winfutcandle5min = winfutcandle5min.sort_values(by='Data', ascending=True)
```

```python
winfutcandle5min
```

```python
# Definindo a coluna 'Data' como índice
winfutcandle5min = winfutcandle5min.set_index('Data')
```

```python
winfutcandle5min
```

```python
winfutcandle5min.plot.line(y="Fechamento", use_index=True)
```

```python
winfutcandle5min["PredictTrade"] = winfutcandle5min["Fechamento"].shift(-1)
```

```python
winfutcandle5min["Alvo"] = (winfutcandle5min["PredictTrade"] > winfutcandle5min["Fechamento"]).astype(int)
```

```python
winfutcandle5min
```

```python
from sklearn.ensemble import RandomForestClassifier

# Creating the RandomForestClassifier model with the desired parameters
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

# Splitting the data into training and test sets
train = winfutcandle5min.iloc[:-100]
test = winfutcandle5min.iloc[-100:]

# Specifying the predictors
predictors = ["Fechamento", "Volume Quantidade", "Abertura", "Máxima", "Mínima"]

# Fitting the model to the training data
model.fit(train[predictors], train["Alvo"])
```

```python
from sklearn.metrics import precision_score

# Making predictions based on the test data
preds = model.predict(test[predictors])

# Creating a pandas series with the predictions and corresponding index
preds = pd.Series(preds, index=test.index)

# Calculating accuracy by comparing the predictions with the actual values
precision_score(test["Alvo"], preds)
```

```python
# Combined DataFrame with the columns "Target" (Alvo) and "Predictions" (Previsões)
combined = pd.concat([test["Alvo"], preds], axis=1)
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Alvo"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Previsoes")
    combined = pd.concat([test["Alvo"], preds], axis=1)
    return combined
```

```python
#108 = number of 5-minute candles in one hour, in one day of trading (9 hours).
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)
```

```python
# Backtest function with the winfutpredict DataFrame, model, and predictors
predictions = backtest(winfutcandle5min, model, predictors)
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = winfutcandle5min.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    winfutcandle5min[ratio_column] = winfutcandle5min["Fechamento"] / rolling_averages ["Fechamento"]

    trend_column = f"Trend_{horizon}"
    winfutcandle5min[trend_column] = winfutcandle5min.shift(1).rolling(horizon).sum()["Alvo"]

    new_predictors+= [ratio_column, trend_column]
```

```python
winfutcandle5min = winfutcandle5min.dropna(subset=winfutcandle5min.columns.difference(["PredictTrade"]))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Alvo"])
    preds = model.predict_proba(test[predictors])[:, 1]
    preds[preds >= 0.6] = 1
    preds[preds < 0.6] = 0
    preds = pd.Series(preds, index=test.index, name="Previsoes")
    combined = pd.concat([test["Alvo"], preds], axis=1)
    return combined
```

```python
# Backtest function with the winfutpredict DataFrame, model, and new predictors
predictions = backtest(winfutcandle5min, model, new_predictors)
```
